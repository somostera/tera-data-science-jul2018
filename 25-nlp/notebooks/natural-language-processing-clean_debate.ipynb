#### Como funciona?

Claro que nada disso √© m√°gica e tudo √© realizado atrav√©s de t√©cnicas e an√°lises l√©xicas, sint√°ticas e sem√¢nticas, al√©m da compreens√£o de quest√µes cognitivas, psicolingu√≠sticas e estat√≠sticas.

Tamb√©m deveria ser capaz de recuperar informa√ß√µes (descobrir quais recursos textuais s√£o necess√°rios), extrair informa√ß√µes (fatos pertinentes a partir destes recursos) e inferir (desenhar conclus√µes baseadas em fatos conhecidos.*

````python
from pylab import rcParams
````

Vamos criar um m√©todo para localizar as ocorr√™ncias de palavras.

````python
def localizar_palavra(agulha, palheiro): 
    return pd.Series(np.histogram(
        [word.i for word in palheiro 
         if word.text.lower() == agulha], bins=50)[0])
````

Vamos buscar por ocorr√™ncias de algumas palavras e plot√°-las.

````python
rcParams['figure.figsize'] = 16, 8

pd.DataFrame(
    {name: localizar_palavra(name.lower(), tweets_spacy) 
     for name in ['bolsonaro', 'marina', 'ciro', 'daciolo', 'alckmin', 'lula', 'haddad', 'boulos', 'meirelles', 
                  'hospital', 'daciolo', 'amoedo']}
).plot(subplots=True)
plt.show()
````

## Vamos brincar com verbos e palavras

Este m√©todo vai buscar verbos (TOP20) relacionados √† uma palavra informada.

````python
def verbs_characters(text, character):
    sents = [sent for sent in text.sents if character in sent.string]
    char_words = []
    for sent in sents: 
        for word in sent: 
            if character in word.string: 
                char_words.append(word)
    char_verbs = []
    for word in char_words: 
        for ancestor in word.ancestors: 
            if ancestor.pos_.startswith('V'): 
                char_verbs.append(ancestor.lemma_.strip())
    return Counter(char_verbs).most_common(20)
````

### Vamos construir ngrams para nossos *tweets*

Este m√©todo criar n-grams de qualquer **doc** spaCy a partir da quantidade **n** informada.

````python
def ngrams(doc, n): 
    doc = [word for word in doc 
           if word.is_alpha 
           if not word.string.isupper()] 
    return list(zip(*[doc[i:] for i in range(n)]))
````

O verbo "estudar", por exemplo, pode ser utilizado em diferentes conjuga√ß√µes al√©m da varia√ß√£o singular/plural:
- Estudo
- Estudei
- Estudaram
- Estudamos
- Etc.

#### Ser√° que interpretar todas as palavras em todas as suas varia√ß√µes √© o melhor cen√°rio de NLP?

N√£o, n√©?

Justamente por isso existem t√©cnicas para "normalizar" as palavras:

- ***Stemming***: talvez a mais conhecida, √© uma t√©cnica respons√°vel por reduzir uma palavra para sua **raiz gramatical**. Ou seja, independente de sua conjuga√ß√£o, a palavra estudar √© sempre reduzida para **estud**. O *stemmer* mais conhecido √© o de **Porter**. Essa t√©cnica √© a utilizada no [NLTK](http://www.nltk.org/howto/stem.html).
- ***Lemmatisation***: ao inv√©s de reduzir uma palavra par sua sua raiz gramatical, esta t√©nica consiste na redu√ß√£o para seu [**lema**](https://priberam.pt/dlpo/lema), ou seja, sua entrada de verbete em dicion√°rios (*e.g.*, "estudei", "estudaram", "estudou", etc, ficariam sempre como **"estudar"**). Essa t√©cnica √© a utilizada pelo [spaCy](https://spacy.io/api/lemmatizer).

````python
lemmas_df['token'] = lemmas_df['token'].apply(lambda x: x.replace('\\n', ''))
````

O m√©todo abaixo normaliza (*lowercase*, tokeniza√ß√£o, lemmetiza√ß√£o) os *tweets*.

````python
def normalizar(comment, lowercase, remove_stopwords):
    if lowercase:
        comment = comment.lower()
    comment = nlp(comment)
    lemmatized = list()
    for word in comment:
        lemma = word.lemma_.strip()
        if lemma:
            if not remove_stopwords or (remove_stopwords and lemma not in STOP_WORDS):
                lemmatized.append(lemma)
    return " ".join(lemmatized)
````

## Classifica√ß√£o supervisionada

Vamos utilizar um algoritmo que voc√™s j√° conhecem e tamb√©m √© muito utilizado:
- √Årvore de Decis√£o *(Random Forest)* 

Importando o que precisamos.

````python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import cross_validate
from sklearn import metrics
````

### √Årvore de Decis√£o

````python
from sklearn.ensemble import RandomForestClassifier
````

Vamos utilizar o conceito de *pipelines* que o scikit nos proporciona.

Nele constru√≠mos um *"workflow"* das atividades que queremos executar ao inv√©s de execut√°-las manualmente uma a uma. No nosso caso:
- Vetoriza√ß√£o de ocorr√™ncias;
- TF-IDF;
- Modelo de classifica√ß√£o supervisionada (DT, NN)

````python
%%time
rf_pl = Pipeline([ ('vect', CountVectorizer()),
                   ('tfidf', TfidfTransformer()),
                   ('clf', RandomForestClassifier())
                 ])
````

Vamos dividir nossa base em **treino** e **teste**.

````python
train_df, test_df = train_test_split(train_set, test_size=0.3)
````

Vamos ocultar a **classe** na base de treino e converter os *dataframes* para **"ndarray"** (input dos algoritmos).

Tamb√©m vamos ocultar a classe da base de teste e guard√°-la para avalia√ß√£o.

````python
X = train_df[train_df.columns[0:len(train_df.axes[1])-1]].values.ravel()
y = train_df.sentiment
y_test = test_df[test_df.columns[0:len(test_df.axes[1])-1]].values.ravel()
````

Agora podemos treinar nossa √Årvore de Decis√£o.

````python
%%time
````

Com o modelo treinado, podemos test√°-lo na base separada de teste sem r√≥tulos:

````python
rf_predicted = rf_pl.predict(y_test)
````

E comparar o resultado com os r√≥tulos corretos que guardamos:

Podemos exibir os resultados atrav√©s das m√©tricas de avalia√ß√£o do modelo: 
- Precis√£o
- Cobertura
- F-measure

````python
print(metrics.classification_report(test_df.sentiment, rf_predicted))
````

Ou a matrix de confus√£o com: 

$\begin{bmatrix}TP & FN\\FP & TN\end{bmatrix}$

Onde:
- **TP**: true positive
- **FN**: false negative
- **FP**: false positive
- **TN**: true negative

````python
print(metrics.confusion_matrix(test_df.sentiment, rf_predicted))
````
